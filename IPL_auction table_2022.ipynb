```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.iplt20.com/auction/2022"
```

```python
req=requests.get(url)
```

```python
req.status_code
```

```python
soup=BeautifulSoup(req.content,"lxml")
```

```python
table=soup.find_all("table",class_="ih-td-tab w-100 auction-tbl")
table = table[0] 
table
```

```python
index=table.find_all("th",class_="skip-filter")
index
```

```python
chinu=[]
for i in index:
    name=i.text
    chinu.append(name)
chinu
```

```python
df=pd.DataFrame(columns=chinu)
```

```python
rows=table.find_all("tr")

for i in rows[1:]:
    data=i.find_all("td")
    row=[tr.text for tr in data]
    first_td=i.find_all("td")[1].find("div",class_="ih-pt-ic").text.strip()
    row[1]=first_td
    

    l=len(df)
    df.loc[l]=row
df    
```
